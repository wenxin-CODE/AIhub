Sigmod

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
import torch
import torch.nn.functional as F

def sigmoid_torch(x):
    return torch.sigmoid(x)
```

Tanh

```python
def tanh(x):
    return np.tanh(x)
```

```python
def tanh_torch(x):
    return torch.tanh(x)
```

ReLU

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def relu_torch(x):
    return F.relu(x)
```

GELU

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(
        np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)
    ))
```

```python
def gelu_torch(x):
    return F.gelu(x)
```

Softmax

```python
def softmax(x, axis=1):
    x = x - np.max(x, axis=axis, keepdims=True)
    exp = np.exp(x)
    return exp / np.sum(exp, axis=axis, keepdims=True)
```

```python
def softmax_torch(x):
    return F.softmax(x, dim=1)
```
